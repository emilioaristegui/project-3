```python
# dependencies
import pandas as pd
from pymongo import MongoClient
from pprint import pprint
```

```python
methane_df = pd.read_csv("Methane_final.csv")
coordinates_df = pd.read_csv("average-latitude-longitude-countries.csv")
merged_df = methane_df.merge(coordinates_df, on="Country")
merged_df.head(5)
```

```python
merged_df.to_csv('merged_df.csv')
```

```python
mongo = MongoClient(port = 27017)
db = mongo.marketing

customer = pd.read_csv("merged_df.csv")
customer = customer.to_dict(orient = 'records')

db.customer_list2.insert_many(customer)
```

```python
results = db.customer_list2.find({'Country':'Mexico'})
for x in results:
    pprint(x)

```
